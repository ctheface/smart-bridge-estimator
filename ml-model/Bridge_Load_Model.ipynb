```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("bridge_data.csv")
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Bridge_ID', 'Location'], inplace=True, errors='ignore')
```

```python
df.drop(columns=['Maintenance_History'], inplace=True, errors='ignore')
df.head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
df["Collapse_Status"].unique()
```

```python
df["Collapse_Status"] = df["Collapse_Status"].map({
    "Collapsed": 1,
    "Standing": 0
})
```

```python
df_encoded = pd.get_dummies(df.drop("Collapse_Status", axis=1), drop_first=True)
X = df_encoded
y = df["Collapse_Status"]
```

```python
import joblib

joblib.dump(df_encoded.columns.tolist(), "collapse_columns.pkl")
```

Training the model on the training data.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_encoded, y, test_size=0.2, random_state=42)

collapse_model = RandomForestClassifier(class_weight="balanced", random_state=42)

collapse_model.fit(X_train, y_train)
```

Testing model on the unseen 20% (X_test):

```python
y_pred = collapse_model.predict(X_test)
```

Performance metrics

```python
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

Saveing the model to a file

```python
import joblib
```

```python

joblib.dump(collapse_model, "bridge_collapse_model.pkl")
```

PHASE 2

```python
y_stress = df["Stress (MPa)"]
```

```python
X_stress = df.drop(columns=["Stress (MPa)", "Strain (%)", "Tensile_Strength (MPa)", "Collapse_Status"])
```

```python
# Turn text columns into numeric columns using one-hot encoding
X_encoded = pd.get_dummies(X_stress, drop_first=True)
```

```python
joblib.dump(X_encoded.columns.tolist(), "model_columns.pkl")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_stress, test_size=0.2, random_state=42
)
```

Regression Model

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(random_state=42)

regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Predict on the test set
y_pred = regressor.predict(X_test)

# Evaluate the prediction accuracy
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
import joblib

joblib.dump(regressor, "stress_model.pkl")
```

Similarly for stress 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
import joblib

y_strain = df["Strain (%)"]

X_strain = df.drop(columns=["Stress (MPa)", "Strain (%)", "Tensile_Strength (MPa)", "Collapse_Status"])
X_strain_encoded = pd.get_dummies(X_strain, drop_first=True)

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_strain_encoded, y_strain, test_size=0.2, random_state=42)

strain_model = RandomForestRegressor(random_state=42)
strain_model.fit(X_train_s, y_train_s)

y_pred_s = strain_model.predict(X_test_s)
print("Strain MAE:", mean_absolute_error(y_test_s, y_pred_s))
print("Strain R²:", r2_score(y_test_s, y_pred_s))

joblib.dump(strain_model, "strain_model.pkl")
```

Similarly for tensile strength

```python

y_tensile = df["Tensile_Strength (MPa)"]

X_tensile = df.drop(columns=["Stress (MPa)", "Strain (%)", "Tensile_Strength (MPa)", "Collapse_Status"])
X_tensile_encoded = pd.get_dummies(X_tensile, drop_first=True)

X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_tensile_encoded, y_tensile, test_size=0.2, random_state=42)

tensile_model = RandomForestRegressor(random_state=42)
tensile_model.fit(X_train_t, y_train_t)

y_pred_t = tensile_model.predict(X_test_t)
print("Tensile Strength MAE:", mean_absolute_error(y_test_t, y_pred_t))
print("Tensile Strength R²:", r2_score(y_test_t, y_pred_t))

joblib.dump(tensile_model, "tensile_model.pkl")
```
